```python
release_dates=pd.read_csv('release_dates.csv')
```

```python
release_dates.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
christmas_movies_usa = release_dates[(release_dates['title'].str.contains('Christmas', case=False)) & (release_dates['country'] == 'USA')]
christmas_movies_usa['date'] = pd.to_datetime(christmas_movies_usa['date'])
christmas_movies_usa['release_month'] = christmas_movies_usa['date'].dt.month
monthly_counts = christmas_movies_usa['release_month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(monthly_counts.index, monthly_counts.values)
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.title('Number of "Christmas" Movies Released in the USA by Month')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
hobbit_movies_usa = release_dates[(release_dates['title'].str.startswith('The Hobbit', na=False)) & (release_dates['country'] == 'USA')]
hobbit_movies_usa['date'] = pd.to_datetime(hobbit_movies_usa['date'])
hobbit_movies_usa['release_month'] = hobbit_movies_usa['date'].dt.month
monthly_counts = hobbit_movies_usa['release_month'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.bar(monthly_counts.index, monthly_counts.values)
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.title('Number of "The Hobbit" Movies Released in the USA by Month')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
romance_movies_usa = release_dates[(release_dates['title'].str.contains('Romance', case=False)) & (release_dates['country'] == 'USA')]
romance_movies_usa['date'] = pd.to_datetime(romance_movies_usa['date'])
romance_movies_usa['release_dayofweek'] = romance_movies_usa['date'].dt.dayofweek
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
romance_movies_usa['release_day'] = romance_movies_usa['release_dayofweek'].map(lambda x: day_names[x])
daily_counts = romance_movies_usa['release_day'].value_counts().reindex(day_names, fill_value=0)
plt.figure(figsize=(10, 6))
daily_counts.plot(kind='bar')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Movies')
plt.title('Number of "Romance" Movies Released in the USA by Day of the Week')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'your_movie_dataset.csv' with the actual file path to your movie dataset in CSV format
file_path = 'C:\Users\user\GitProjects\AIMLRepo\Python-Practices\Python_Day10_PS'

# Read the CSV file into a DataFrame
movies_df = pd.read_csv(file_path)

# Filter movies with Tom Cruise in the cast and released in the USA
tom_cruise_movies_usa = movies_df[(movies_df['Cast'].str.contains('Tom Cruise', case=False, na=False)) & (movies_df['Country'] == 'USA')]

# Extract the month from the release date (assuming 'Release Date' is in 'YYYY-MM-DD' format)
tom_cruise_movies_usa['Release Month'] = pd.to_datetime(tom_cruise_movies_usa['Release Date']).dt.month

# Count the number of movies released in each month
month_counts = tom_cruise_movies_usa['Release Month'].value_counts().sort_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
month_counts.plot(kind='bar', color='skyblue')
plt.title('Release Months of Tom Cruise Movies in the USA')
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
