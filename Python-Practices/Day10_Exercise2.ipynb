```python
titles = pd.read_csv('titles.csv', index_col=None)
```

```python
cast = pd.read_csv('cast.csv', index_col=None)
```

```python
top_10_common_titles = titles['title'].value_counts().head(10)

print(top_10_common_titles)
```

```python
titles_1930s = titles[(titles['year'] >= 1930) & (titles['year'] <= 1939)]
film_counts_by_year = titles_1930s['year'].value_counts()
top_3_years = film_counts_by_year.head(3)
print(top_3_years)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'titles' with a 'year' column

# Create a new column 'decade' by rounding down the 'year' to the nearest decade
titles['decade'] = titles['year'] // 10 * 10

# Group the DataFrame by 'decade' and count the number of films for each decade
film_counts_by_decade = titles['decade'].value_counts().sort_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
film_counts_by_decade.plot(kind='bar', color='skyblue')
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.title('Number of Films Released Each Decade')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'titles' with a 'year' column and a 'title' column

# Filter the DataFrame to include only "Hamlet" films
hamlet_films = titles[titles['title'] == 'Hamlet']

# Create a new column 'decade' by rounding down the 'year' to the nearest decade
hamlet_films['decade'] = hamlet_films['year'] // 10 * 10

# Group the DataFrame by 'decade' and count the number of "Hamlet" films for each decade
hamlet_counts_by_decade = hamlet_films['decade'].value_counts().sort_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
hamlet_counts_by_decade.plot(kind='bar', color='skyblue')
plt.xlabel('Decade')
plt.ylabel('Number of "Hamlet" Films Released')
plt.title('Number of "Hamlet" Films Released Each Decade')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame 'cast' with 'character' and 'year' columns

# Filter the DataFrame to include only rows where 'character' is "Rustler"
rustler_characters = cast[cast['character'] == 'Rustler']

# Create a new column 'decade' by rounding down the 'year' to the nearest decade
rustler_characters['decade'] = rustler_characters['year'] // 10 * 10

# Group the DataFrame by 'decade' and count the number of "Rustler" characters for each decade
rustler_counts_by_decade = rustler_characters['decade'].value_counts().sort_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
rustler_counts_by_decade.plot(kind='bar', color='skyblue')
plt.xlabel('Decade')
plt.ylabel('Number of "Rustler" Characters')
plt.title('Number of "Rustler" Characters in Each Decade')
plt.xticks(rotation=45)
plt.show()
```
