```python
#Create a class for quizzes and questions
import math

class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Get user input for shape dimensions
def get_circle_input():
    radius = float(input("Enter the radius of the circle: "))
    return Circle(radius)

def get_rectangle_input():
    length = float(input("Enter the length of the rectangle: "))
    width = float(input("Enter the width of the rectangle: "))
    return Rectangle(length, width)

# Demonstrate the usage of the classes with user input
if __name__ == "__main__":
    choice = input("Choose a shape (circle or rectangle): ").lower()

    if choice == "circle":
        shape = get_circle_input()
    elif choice == "rectangle":
        shape = get_rectangle_input()
    else:
        print("Invalid choice")
        exit()

    print(f"{choice.capitalize()} - Area: {shape.area()}, Perimeter: {shape.perimeter()}")
```

```python
#calculates the user's score on a quiz.
class Question:
    def __init__(self, text, options, correct_option):
        self.text = text
        self.options = options
        self.correct_option = correct_option

class Quiz:
    def __init__(self, name, questions):
        self.name = name
        self.questions = questions

    def calculate_score(self, user_answers):
        if len(user_answers) != len(self.questions):
            raise ValueError("Invalid number of answers provided")

        score = 0
        for user_answer, question in zip(user_answers, self.questions):
            if user_answer == question.correct_option:
                score += 1
        return score

# Get user input for quiz answers
def get_user_answers(quiz):
    user_answers = []
    for i, question in enumerate(quiz.questions):
        print(f"Question {i + 1}: {question.text}")
        for j, option in enumerate(question.options):
            print(f"{j + 1}. {option}")
        answer = int(input("Enter your answer (1, 2, 3, 4): "))
        user_answers.append(answer)
    return user_answers

# Example usage with user input:
if __name__ == "__main__":
    question1 = Question("What is the capital of France?", ["London", "Berlin", "Paris", "Madrid"], 2)
    question2 = Question("Which gas do plants absorb?", ["Oxygen", "Carbon Dioxide", "Nitrogen", "Hydrogen"], 1)
    question3 = Question("What is 2 + 2?", ["3", "4", "5", "6"], 2)

    quiz = Quiz("General Knowledge Quiz", [question1, question2, question3])

    user_answers = get_user_answers(quiz)

    user_score = quiz.calculate_score(user_answers)
    print(f"User Score: {user_score}")
```

```python
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance):
        self.__account_number = account_number  # Private attribute for account number
        self.__account_holder_name = account_holder_name  # Private attribute for account holder's name
        self.__balance = balance  # Private attribute for account balance

    # Getter methods to access private attributes
    def get_account_number(self):
        return self.__account_number

    def get_account_holder_name(self):
        return self.__account_holder_name

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        try:
            amount = float(amount)
            if amount > 0:
                self.__balance += amount
                print(f"Deposited ${amount}. New balance: ${self.__balance}")
            else:
                print("Invalid deposit amount. Please enter a positive value.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

    def withdraw(self, amount):
        try:
            amount = float(amount)
            if 0 < amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrew ${amount}. New balance: ${self.__balance}")
            else:
                print("Invalid withdrawal amount or insufficient balance.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance, interest_rate):
        super().__init__(account_number, account_holder_name, balance)
        self.__interest_rate = interest_rate  # Private attribute for interest rate

    def calculate_interest(self):
        interest = (self.__interest_rate / 100) * self.get_balance()  # Calculate interest based on balance and interest rate
        self.deposit(interest)  # Add interest to the balance using the deposit method
        print(f"Interest of ${interest} added. New balance: ${self.get_balance()}")

class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance, overdraft_limit):
        super().__init__(account_number, account_holder_name, balance)
        self.__overdraft_limit = overdraft_limit  # Private attribute for overdraft limit

    # Override the withdraw method to allow withdrawals within overdraft limit
    def withdraw(self, amount):
        try:
            amount = float(amount)
            available_balance = self.get_balance() + self.__overdraft_limit  # Calculate available balance with overdraft
            if 0 < amount <= available_balance:
                self._BankAccount__balance -= amount  # Use the base class's private attribute with name mangling
                print(f"Withdrew ${amount}. New balance: ${self.get_balance()}")
            else:
                print("Invalid withdrawal amount or exceeds overdraft limit.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# Example usage:
if __name__ == "__main__":
    savings_account = SavingsAccount("12345", "John Doe", 1000, 5.0)
    checking_account = CheckingAccount("67890", "Alice Johnson", 1500, 500)

    print("Savings Account:")
    savings_account.deposit(500)
    savings_account.calculate_interest()
    savings_account.withdraw(200)

    print("\nChecking Account:")
    checking_account.deposit(300)
    checking_account.withdraw(2200)
```

```python
# Base class "Employee"
class Employee:
    def __init__(self, name, employee_id, salary):
        self.__name = name
        self.__employee_id = employee_id
        self.__salary = salary

    # Getter methods
    def get_name(self):
        return self.__name

    def get_employee_id(self):
        return self.__employee_id

    def get_salary(self):
        return self.__salary

    # Setter methods
    def set_name(self, name):
        self.__name = name

    def set_employee_id(self, employee_id):
        self.__employee_id = employee_id

    def set_salary(self, salary):
        self.__salary = salary

    # Method to calculate salary (to be overridden by subclasses)
    def calculate_salary(self):
        pass

# Subclass "FullTimeEmployee"
class FullTimeEmployee(Employee):
    def __init__(self, name, employee_id, annual_salary):
        super().__init__(name, employee_id, annual_salary)

    # Override the salary calculation method for full-time employees
    def calculate_salary(self):
        return self.get_salary()

# Subclass "PartTimeEmployee"
class PartTimeEmployee(Employee):
    def __init__(self, name, employee_id, hourly_rate, hours_worked):
        super().__init__(name, employee_id, None)
        self.__hourly_rate = hourly_rate
        self.__hours_worked = hours_worked

    # Getter methods for hourly_rate and hours_worked
    def get_hourly_rate(self):
        return self.__hourly_rate

    def get_hours_worked(self):
        return self.__hours_worked

    # Setter methods for hourly_rate and hours_worked
    def set_hourly_rate(self, hourly_rate):
        self.__hourly_rate = hourly_rate

    def set_hours_worked(self, hours_worked):
        self.__hours_worked = hours_worked

    # Override the salary calculation method for part-time employees
    def calculate_salary(self):
        return self.get_hourly_rate() * self.get_hours_worked()

# Input from the user for Full-Time Employee
name = input("Enter Full-Time Employee Name: ")
employee_id = int(input("Enter Employee ID: "))
annual_salary = float(input("Enter Annual Salary: "))

# Create an instance of FullTimeEmployee
full_time_employee = FullTimeEmployee(name, employee_id, annual_salary)

# Input from the user for Part-Time Employee
name = input("\nEnter Part-Time Employee Name: ")
employee_id = int(input("Enter Employee ID: "))
hourly_rate = float(input("Enter Hourly Rate: "))
hours_worked = float(input("Enter Hours Worked: "))

# Create an instance of PartTimeEmployee
part_time_employee = PartTimeEmployee(name, employee_id, hourly_rate, hours_worked)

# Display employee information
print("\nFull-Time Employee Information:")
print(f"{full_time_employee.get_name()} (Employee ID: {full_time_employee.get_employee_id()})")
print(f"Annual Salary: ${full_time_employee.calculate_salary()}")

print("\nPart-Time Employee Information:")
print(f"{part_time_employee.get_name()} (Employee ID: {part_time_employee.get_employee_id()})")
print(f"Hourly Rate: ${part_time_employee.get_hourly_rate()}")
print(f"Hours Worked: {part_time_employee.get_hours_worked()}")
print(f"Monthly Salary: ${part_time_employee.calculate_salary()}")
```

```python
class Book:
    def __init__(self, book_id, title, author, publication_year, is_available=True):
        self.__book_id = book_id
        self.__title = title
        self.__author = author
        self.__publication_year = publication_year
        self.__is_available = is_available

    # Getter and setter methods (omitted for brevity)

class Fiction(Book):
    def __init__(self, book_id, title, author, publication_year, genre):
        super().__init__(book_id, title, author, publication_year)
        self.__genre = genre

    # Getter and setter methods (omitted for brevity)

class NonFiction(Book):
    def __init__(self, book_id, title, author, publication_year, topic):
        super().__init__(book_id, title, author, publication_year)
        self.__topic = topic

    # Getter and setter methods (omitted for brevity)

class Patron:
    def __init__(self, patron_id, name, email):
        self.__patron_id = patron_id
        self.__name = name
        self.__email = email

    # Getter and setter methods (omitted for brevity)

class Transaction:
    def __init__(self, transaction_id, book, patron, due_date):
        self.__transaction_id = transaction_id
        self.__book = book
        self.__patron = patron
        self.__due_date = due_date
        self.__return_date = None

    # Getter and setter methods (omitted for brevity)

class Library:
    def __init__(self):
        self.__books = []
        self.__patrons = []
        self.__transactions = []

    def add_book(self, book):
        self.__books.append(book)

    def add_patron(self, patron):
        self.__patrons.append(patron)

    def check_out_book(self, book, patron, due_date):
        if book.is_available():
            transaction_id = len(self.__transactions) + 1
            transaction = Transaction(transaction_id, book, patron, due_date)
            self.__transactions.append(transaction)
            book.set_is_available(False)
            return transaction
        else:
            return None

    def return_book(self, transaction):
        transaction.get_book().set_is_available(True)
        transaction.set_return_date("today")

    def track_overdue_books(self):
        overdue_books = []
        for transaction in self.__transactions:
            if transaction.get_return_date() is None:
                overdue_books.append(transaction)
        return overdue_books

    def notify_patrons(self, overdue_books):
        for transaction in overdue_books:
            patron = transaction.get_patron()
            print(f"Dear {patron.get_name()}, your book '{transaction.get_book().get_title()}' is overdue. Please return it.")

# Example usage with user input:
library = Library()

while True:
    print("\nLibrary Management Menu:")
    print("1. Add Book")
    print("2. Add Patron")
    print("3. Check Out Book")
    print("4. Return Book")
    print("5. List Overdue Books")
    print("6. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        book_id = int(input("Enter Book ID: "))
        title = input("Enter Title: ")
        author = input("Enter Author: ")
        publication_year = int(input("Enter Publication Year: "))
        genre_or_topic = input("Enter Genre/Topic: ")
        is_fiction = input("Is it Fiction (yes/no): ").lower()
        if is_fiction == "yes":
            book = Fiction(book_id, title, author, publication_year, genre_or_topic)
        else:
            book = NonFiction(book_id, title, author, publication_year, genre_or_topic)
        library.add_book(book)
        print("Book added successfully!")

    elif choice == "2":
        patron_id = int(input("Enter Patron ID: "))
        name = input("Enter Patron Name: ")
        email = input("Enter Patron Email: ")
        patron = Patron(patron_id, name, email)
        library.add_patron(patron)
        print("Patron added successfully!")

    elif choice == "3":
        book_id = int(input("Enter Book ID to Check Out: "))
        patron_id = int(input("Enter Patron ID: "))
        due_date = input("Enter Due Date (YYYY-MM-DD): ")
        book = None
        patron = None

        # Find the book and patron by their IDs
        for b in library._Library__books:
            if b.get_book_id() == book_id:
                book = b
                break

        for p in library._Library__patrons:
            if p.get_patron_id() == patron_id:
                patron = p
                break

        if book and patron:
            transaction = library.check_out_book(book, patron, due_date)
            if transaction:
                print(f"Book '{book.get_title()}' checked out successfully to {patron.get_name()}.")
            else:
                print("Book is not available for checkout.")
        else:
            print("Invalid Book ID or Patron ID.")

    elif choice == "4":
        transaction_id = int(input("Enter Transaction ID to Return Book: "))
        transaction = None

        # Find the transaction by its ID
        for t in library._Library__transactions:
            if t.get_transaction_id() == transaction_id:
                transaction = t
                break

        if transaction:
            library.return_book(transaction)
            print(f"Book '{transaction.get_book().get_title()}' returned successfully.")
        else:
            print("Invalid Transaction ID.")

    elif choice == "5":
        overdue_books = library.track_overdue_books()
        if overdue_books:
            print("\nOverdue Books:")
            for transaction in overdue_books:
                print(f"{transaction.get_book().get_title()} (Due Date: {transaction.get_due_date()})")
        else:
            print("No overdue books.")

    elif choice == "6":
        print("Exiting Library Management System.")
        break

    else:
        print("Invalid choice. Please try again.")
```

```python
class Product:
    def __init__(self, product_id, name, price, description, stock):
        self.__product_id = product_id
        self.__name = name
        self.__price = price
        self.__description = description
        self.__stock = stock

    def get_product_id(self):
        return self.__product_id

    def get_name(self):
        return self.__name

    def get_price(self):
        return self.__price

    def get_description(self):
        return self.__description

    def get_stock(self):
        return self.__stock

    def set_price(self, price):
        self.__price = price

    def set_description(self, description):
        self.__description = description

    def set_stock(self, stock):
        self.__stock = stock

    def __str__(self):
        return f"{self.__name} - ${self.__price}"

class Electronics(Product):
    def __init__(self, product_id, name, price, description, stock, warranty):
        super().__init__(product_id, name, price, description, stock)
        self.__warranty = warranty

    def get_warranty(self):
        return self.__warranty

    def set_warranty(self, warranty):
        self.__warranty = warranty

    def __str__(self):
        return f"{super().__str__()} (Warranty: {self.__warranty} months)"

class Clothing(Product):
    def __init__(self, product_id, name, price, description, stock, size):
        super().__init__(product_id, name, price, description, stock)
        self.__size = size

    def get_size(self):
        return self.__size

    def set_size(self, size):
        self.__size = size

    def __str__(self):
        return f"{super().__str__()} (Size: {self.__size})"

class ShoppingCart:
    def __init__(self):
        self.__items = []

    def add_item(self, product, quantity):
        if product.get_stock() >= quantity:
            self.__items.append((product, quantity))
            product.set_stock(product.get_stock() - quantity)
            return True
        else:
            return False

    def remove_item(self, product, quantity):
        for item in self.__items:
            if item[0] == product:
                if item[1] >= quantity:
                    item[1] -= quantity
                    product.set_stock(product.get_stock() + quantity)
                    if item[1] == 0:
                        self.__items.remove(item)
                    return True
                else:
                    return False
        return False

    def calculate_total_cost(self):
        total_cost = 0
        for item in self.__items:
            total_cost += item[0].get_price() * item[1]
        return total_cost

    def view_cart(self):
        for item in self.__items:
            product, quantity = item
            print(f"{product.get_name()} ({product.get_price()} x {quantity})")

class Order:
    def __init__(self, order_id, customer_name, cart):
        self.__order_id = order_id
        self.__customer_name = customer_name
        self.__cart = cart

    def place_order(self):
        # Implement order placement logic here (e.g., payment processing)
        pass

    def get_order_summary(self):
        return f"Order ID: {self.__order_id}\nCustomer: {self.__customer_name}\nTotal Cost: ${self.__cart.calculate_total_cost()}"

# Example usage with user input:
electronics_product = Electronics(1, "Laptop", 800, "High-performance laptop", 10, 24)
clothing_product = Clothing(2, "T-shirt", 20, "Cotton T-shirt", 50, "M")

cart = ShoppingCart()

while True:
    print("\n1. Add product to cart")
    print("2. Remove product from cart")
    print("3. View cart")
    print("4. Checkout and place order")
    print("5. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        print("Available Products:")
        print("1. Laptop - $800")
        print("2. T-shirt - $20")
        product_choice = input("Enter the product number: ")
        quantity = int(input("Enter the quantity: "))
        if product_choice == "1":
            cart.add_item(electronics_product, quantity)
        elif product_choice == "2":
            cart.add_item(clothing_product, quantity)
        else:
            print("Invalid product choice.")

    elif choice == "2":
        cart.view_cart()
        product_choice = input("Enter the product number to remove: ")
        quantity = int(input("Enter the quantity to remove: "))
        if product_choice == "1":
            cart.remove_item(electronics_product, quantity)
        elif product_choice == "2":
            cart.remove_item(clothing_product, quantity)
        else:
            print("Invalid product choice.")

    elif choice == "3":
        print("\nShopping Cart Contents:")
        cart.view_cart()

    elif choice == "4":
        if cart.calculate_total_cost() > 0:
            customer_name = input("Enter your name: ")
            order = Order(1001, customer_name, cart)
            print("\nOrder Summary:")
            print(order.get_order_summary())
            order.place_order()
            print("Order placed successfully!")
            break
        else:
            print("Your cart is empty. Please add items to your cart first.")

    elif choice == "5":
        print("Thank you for shopping with us. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```
